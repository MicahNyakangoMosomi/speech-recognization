```python
import wave
import numpy as np
import matplotlib.pyplot as plt
```

```python
obj = wave.open("./BAK.wav","rb")
```

```python
print(obj.getparams())
```

```python
time = obj.getnframes()/obj.getframerate()
print(time)
```

```python
frames = obj.readframes(-1)
```

```python
print(len(frames))
obj.close()
```

```python
signal_array = np.frombuffer(frames,dtype=np.int16)
signal_array = signal_array.reshape(-1,2)
left = signal_array[:,0]
right = signal_array[:,1]
t_signal = np.linspace(0,time,len(right))
t_signal.shape
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(t_signal,left)
ax[0].set_title("Left Channel")
ax[1].plot(t_signal,right)
ax[1].set_title("Right Channel")
plt.show()
```

```python
obj_new = wave.open("newBAK.wav","wb")
obj_new.setnchannels(2)
obj_new.setsampwidth(2)
obj_new.setframerate(44100)
obj_new.writeframes(frames)
obj_new.close
```
